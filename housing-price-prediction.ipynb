We will remove the data points that fall outside the mean and std per location.

```python
def remove_pps_outliers(data):
    data_out = pd.DataFrame()
    for key, subdata in data.groupby('Location'):
        m = np.mean(subdata['Price/SqFt'])
        st = np.std(subdata['Price/SqFt'])
        reduced_data = subdata[(subdata['Price/SqFt']>(m-st)) & (subdata['Price/SqFt']<=(m+st))]
        data_out = pd.concat([data_out,reduced_data],ignore_index=True)
    return data_out    
```

```python
data4 = remove_pps_outliers(data3)
data4.shape
```

We will remove the data points whose prices per sqft. fall outside the mean and std per location.

```python
data4 = remove_pps_outliers(data3)
data4.rename(columns={"Price(in Lakhs)": "Price"}, inplace=True)
```

```python
import matplotlib
def plot_scatter_chart(df,location):
    bhk2 = df[(df.Location==location) & (df.BHK==2)]
    bhk3 = df[(df.Location==location) & (df.BHK==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10) #This line sets the default figure size for the plot to a 
    #width of 15 inches and a height of 10 inches, ensuring that the plot is large and easy to read.
    plt.scatter(bhk2.Total_SqFt,bhk2.Price,color='blue',label='2 BHK', s=50)
    plt.scatter(bhk3.Total_SqFt,bhk3.Price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(location)
    plt.legend()
```

From the above we find that there are some 3BHK's whose prices are found to be below the prices of 2BHK's at the same locality with both the 2BHK's and the 3BHK's having the same built up area, so we will have to clean these respective data points.

Similarly, Now we can remove those 2 BHK apartments whose price_per_sqft is less than mean price_per_sqft of 1 BHK apartment.

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('Location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('BHK'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df['Price/SqFt']),
                'std': np.std(bhk_df['Price/SqFt']),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('BHK'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df['Price/SqFt']<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
```

```python
data5 = remove_bhk_outliers(data4)
data5.shape
```

Dividing our Model into Train_Test Split


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Define multiple models in a dictionary
multiple_models = {
    "DecisionTreeRegressor": DecisionTreeRegressor(),
    "RandomForestRegressor": RandomForestRegressor(),
    "LinearRegression": LinearRegression()    
}

# Loop over models, fit them, and evaluate the score
for model_name, model in multiple_models.items():
    model.fit(X_train, y_train)  # Fit each model
    score = model.score(X_test, y_test)  # Evaluate the model
    print(f"{model_name} Score: {score}")
```

As we see from the above that Linear Regression has more accuracy than the other Regressors.

```python
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

Test the model for few properties

```python
import warnings

def predict_price(location, sqft, bath, bhk):
    try:
        # Check if location exists in the dataset
        loc_index = np.where(X.columns == location)[0][0]

        # Create an empty array for the feature vector
        x = np.zeros(len(X.columns))
        x[0] = sqft
        x[1] = bath
        x[2] = bhk

        # Set the location one-hot encoding if valid
        if loc_index >= 0:
            x[loc_index] = 1

        # Suppress specific warnings using warnings.catch_warnings()
        with warnings.catch_warnings():
            warnings.simplefilter("ignore", UserWarning)  # Ignore only UserWarnings
            # Make the prediction
            predicted_price = round(float(lr_clf.predict([x])[0]),2)

        return "Rs", predicted_price, "Lakhs"

    except IndexError:
        # Handle the case where the location is not found
        return "Error: Location not found."

```
