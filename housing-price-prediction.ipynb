```python
data["bhk"] = data["size"].apply(lambda x: int(x.split(' ')[0]))
```

```python
data.drop(["size"],axis = "columns")
```

```python
data["bhk"].unique()
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens)== 2:
        return(float(tokens[0])+float(tokens[1]))/2 
    try:
        return float(x)
    except:
        return None
```

```python
data2 = data.copy()
data2["total_sqft"] = data2["total_sqft"].apply(convert_sqft_to_num)
data2
```

```python
data1 = data2.copy()
data1['price_per_sqft'] = data1["price"]*100000/data1["total_sqft"]
data1.head()
```

```python
data1.drop(['size'], axis = "columns", inplace = True)
```

```python
data1
```

```python
data1.rename(columns={'price':"Price(in Lakhs)", 'location' : "Location", 'total_sqft': "Total_SqFt", 'bath' : "Bath", "balcony":"Balcony", "bhk" : "BHK's", 'price_per_sqft':"Price/SqFt"}, inplace = True)
```

```python
len(data1['Total_SqFt']=="NaN")
```

```python
data1.Location.unique()
```

```python
len(data1.Location.unique())
```

```python
data1.location = data1.Location.apply(lambda x : x.strip())
location_stats = data1.groupby('Location')['Location'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_stats_less_than_10 = location_stats[location_stats<=10]
location_stats_less_than_10
```

```python
len(data1.Location.unique())
```

```python
data1.Location = data1.Location.apply(lambda x : 'Other' if x in location_stats_less_than_10 else x)
len(data1.Location.unique())
```
